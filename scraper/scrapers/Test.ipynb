```python
from scraper.scrapers import Scraper
from scraper.scrapers.registry import _registry
from scraper.scrapers.base import ScrapeResult
from scraper.utils.scrapers.mixins import (
    ChromeDriverProvider,
    ClientMixin,
    BeautifulSoupMixin,
)

_registry.pop("test", None)

class TestScraper(Scraper, ChromeDriverProvider, ClientMixin, BeautifulSoupMixin):
    
    scraper_name = "test"
    
    def scrape(self):
        driver = self.get_remote_driver()
        url = self.resource.url
        driver.get(url)
        driver.quit()
        scraped_data = self.build_scraped_data({"a": "b"})
        return ScrapeResult(data=scraped_data, state={})
```

```python
scraper = TestScraper()
```

```python
driver = scraper.get_remote_driver()
```
